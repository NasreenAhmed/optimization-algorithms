```python
def f(x):
   return (x-2)**2

def f_der(f,x):
  h = 0.00001
  top = f(x+h)-f(x)
  bottom = h
  slope = top/bottom
  return float("%.2f"% slope)
```

```python
def bisection_root_finding(a,b,eps):

  if a > b : 
    a,b = b,a 
  
  f_a, f_b = f_der(f,a), f_der(f,b)

  print("a:",a,"b",b,"f_a:",f_a,"f_b:",f_b)
  if f_a == 0:
    b = a
    print("a:",a,"b",b,"f_a:",f_a,"f_b:",f_b)
    return (a)

  if f_b == 0:
    a = b
    #print("1","a:",a,"b",b,"f_a:",f_a,"f_b:",f_b)
    return (b)
  iter = 0
  print("2","a:",a,"b",b,"f_a:",f_a,"f_b:",f_b)
  while abs(b - a) > eps:
    iter +=1

    x = (a+b)/2

    print("Value of x:",x)

    y = f_der(f,x)

    print("Value of y:",y)

    if y == 0:
      a,b = x, x
      break

    print("3","a:",a,"b",b,"f_a:",f_a,"f_b:",f_b)

    if np.sign(y) == np.sign(f_a):
      a = x
    else:
      b = x
    print("iter:",iter,"a:",a,"b",b)
  return (a+b)/2
```

```python
a = -30
b = 5
eps = 1e-5
bisection_root_finding(a,b,eps)
```
